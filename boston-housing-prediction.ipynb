```python
# Importing modules
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Selecting our X feature(s) (Independent features)
```

```python
X = df[['RM', 'PTRATIO', 'LSTAT']]
```

```python
X.shape
```

```python
# Selecting our Y feature (Dependent/Target Feature)
```

```python
y = df['PRICE']
y.shape
```

```python
y = np.array(y).reshape(-1,1)
```

```python
y.ndim
```

##### *APPLYING POLYNOMIAL FEATURING, FEATURE SCALING, CROSS VALIDATION, REGULARIZATION AND INITIATING OUR LINEAR REGRESSION MODEL*

```python
# Instantiating objects

poly = PolynomialFeatures(degree=2)
scaler = StandardScaler()
model = Ridge(alpha=0.1)

# Applying K-Fold CV
kf = KFold(n_splits=5, shuffle=True, random_state=5)
r2 = []
mse_score = []

for train_idx, test_idx in kf.split(X):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    #Polynomial Featuring
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)

    #Scaling Features
    X_train_scaled = scaler.fit_transform(X_train_poly)
    X_test_scaled = scaler.transform(X_test_poly)

    
    #Ridge Regression (L2 Regularization)
    
    model.fit(X_train_scaled, y_train)

    y_pred = model.predict(X_test_scaled)

    r2.append(r2_score(y_test, y_pred))
    
    mse_score.append(mean_squared_error(y_test, y_pred))
    

avg_kf = np.mean(r2)
print(f"Average R2 score = {avg_kf * 100:.2f}")

avg_mse = np.mean(mse_score)
print(f'Average MSE = {avg_mse:.2f}')
```

#### *Plotting Actual Price vs Predicted Price*

```python
plt.figure(figsize=(10,5))
plt.scatter(y_test, y_pred, alpha=0.7, color='blue')  
plt.xlabel("Actual House Prices ($1000)")
plt.ylabel("Predicted House Prices ($1000)")
plt.title("Actual vs Predicted Prices")

# Diagonal reference line
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'k--', lw=2)

plt.text(x=min(y_test), 
         y=max(y_pred), 
         s=f'R Score = {avg_kf*100:.3f}', 
         fontsize=10, 
         bbox=dict(facecolor='white', edgecolor='black'))

plt.grid(True)
plt.show()
```

#### *Testing our model on new/unseen data*
